```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

class RegresionLogistica: 

    """Regresion logisitica binaria con descenso por gradiente y entropia cruzada como funcion de costo """
    

    def __init__(self) -> None: 
        self.params = np.empty(0)  
        self.cost_history = []     
        self.params_history = []   

    @staticmethod
    def sigmoide(z: np.ndarray) -> np.ndarray:

        """
        Definición de la función sigmoide para regresión logística binaria.
        :param z: conjunto de entrenamiento 
        :type z: np.ndarray
        :return: probabilidad de que la clase sea 1
        :type return: np.ndarray
        :authors: Camila Gonzalez, Luciana Parallela.
        :date: 25/06/2026
        """
        return np.where( 
        z >= 0,
        1 / (1 + np.exp(-z)),
        np.exp(z) / (1 + np.exp(z)))
    
     
    def fit(self, x_train:np.ndarray, y_train:np.ndarray, max_iter:int = 1000000, convergence_rate: float = 0.00001, alpha = 0.0001) -> None:
        
        """ 
        Entrenamiento de la regresion logistica
        
        :param x_train: conjunto de entrenamiento
        :type x_train:np.ndarray
        :param y_train: labels de entrenamiento
        :type y_train: np.ndarray
        :param max_iter: cantidad maxima de iteraciones antes de detener el descenso por gradiente
        :type max_iter: int
        :param convergence_rate: tasa de convergencia para detener el descenso por gradiente
        :type convergence_rate: float
        :param alpha: tasa de aprendizaje 
        :type alpha: float
        :return: None
        :type return: None
        :authors: Camila Gonzalez, Luciana Parallela.
        :date: 25/06/2026
        """
        y_train = y_train.ravel()
        
        x = np.hstack((np.ones((x_train.shape[0], 1)), x_train))
        
        cost = []
        self.params = np.zeros(x.shape[1]) 
        contador = 0 
        while contador < max_iter:
            contador += 1
            predicciones = np.array([self.sigmoide(np.dot(i, self.params.T)) for i in x])
        
            cost.append(self.funcion_costo(y_train, predicciones))
            vector_vertical = predicciones - y_train 
            gradiente = np.dot(x.T, vector_vertical)       
            self.params_history.append(self.params.copy())
            old_params = self.params.copy()
            self.params = old_params - alpha*gradiente
        
            if np.linalg.norm(self.params - old_params) <= convergence_rate: 
                break                                                         
        
        self.cost_history = cost

    def funcion_costo(self, X: np.ndarray, y_true: np.ndarray) -> float:
        """"
        Funcion de costo para regresion logistica binearia mediante la entropia cruzada

        :param X: conjunto de entrenamiento
        :type X: np.ndarray
        :param y: labels de entrenamiento
        :type y: np.ndarray
        :return: valor promedio de la funcion de costo 
        :type return: float
        :authors: Camila Gonzalez, Luciana Parallela
        :date: 25/06/2026

        """
        probabilidad = self.sigmoide(X) 
    
        loss = -np.mean(
        y_true * np.log(probabilidad) + (1 - y_true) * np.log(1 - probabilidad))
    
        return loss 
       

    def plot_cost_history(self) -> None:

        """"
        Gráfico de la caída de la función de costo a lo largo de las iteraciones del descenso por gradiente. 
        :return: None
        :type return: None
        :authors: Camila Gonzalez, Luciana Parallela
        :date: 25/06/2026
        """
        plt.plot(self.cost_history)
        plt.xlabel('Iteracion')
        plt.ylabel('Costo')
        plt.title('Caida de la funcion de costo') 
        plt.show()

    
```
